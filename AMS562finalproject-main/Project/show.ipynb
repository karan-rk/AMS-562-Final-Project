```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import interact, IntSlider

# File path to your data file
file_path = '/Users/Karan/Desktop/Project/simulation_output.csv'  # Update this with your file path

# Read the data from the file using Pandas
data = pd.read_csv(file_path)

# Function to update the plot dynamically
def update_plot(line_index):
    # Fetch the row for the given line index
    row = data.iloc[line_index]

    # Create figure and axis objects
    fig, ax = plt.subplots()
    categories = ['Susceptible', 'Infected', 'Recovered', 'Vaccinated', 'Dead', 'NewInfections']
    ax.bar(np.arange(len(categories)), row[categories])
    ax.set_xlabel('Categories')
    ax.set_ylabel('Values')
    ax.set_title(f'Chart for Time Step {line_index}')
    ax.set_xticks(np.arange(len(categories)))
    ax.set_xticklabels(categories)
    ax.set_ylim(0, row[categories].max() * 1.1)  # Adjust the y-axis limit dynamically
    
    plt.show()

# Adjust the slider range
line_slider = IntSlider(min=0, max=len(data) - 1, step=1, value=0, description='Time Step')
interact(update_plot, line_index=line_slider)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load the dataset
file_path = '/Users/Karan/Desktop/Project/simulation_output.csv'  # Update with the actual path
data = pd.read_csv(file_path)

# Assuming 'Infected' is the column we're interested in
infected = data['Infected']

# Fit the ARIMA model
model = ARIMA(infected, order=(5, 1, 0))  # You might need to adjust the order
model_fit = model.fit()

# Forecast the next 10 time steps
forecast = model_fit.forecast(steps=10)

# Plotting the forecast
plt.figure(figsize=(10, 6))
plt.plot(infected, label='Infected')
plt.plot(range(len(infected), len(infected) + 10), forecast, label='Forecast', color='red')
plt.xlabel('Time Step')
plt.ylabel('Infected Individuals')
plt.title('Infected Individuals Forecast')
plt.legend()
plt.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
file_path = '//Users/Karan/Desktop/Project/simulation_output.csv'  # Update with your file path
data = pd.read_csv(file_path)

# Choose the dimensions to plot
x_dimension = 'Infected'
y_dimension = 'Recovered'
z_dimension = 'Susceptible'
color_dimension = 'NewInfections'  # Additional dimension for color

# Create a 3D scatter plot with enhanced features
fig = px.scatter_3d(
    data, 
    x=x_dimension, 
    y=y_dimension, 
    z=z_dimension, 
    color=color_dimension,  # Coloring based on an additional metric
    size='Dead',            # Size based on another metric (e.g., 'Dead')
    size_max=10,            # Maximum marker size
    opacity=0.7             # Marker opacity
)

# Update the layout and axes titles
fig.update_layout(
    title='Enhanced 3D Plot of Simulation Data',
    scene=dict(
        xaxis_title=x_dimension,
        yaxis_title=y_dimension,
        zaxis_title=z_dimension
    ),
    legend_title=color_dimension,
    margin=dict(l=0, r=0, b=0, t=0)  # Adjust margins
)

# Show the plot
fig.show()
```
